4 Remove stop words from the cleansed text, what are the stop words? 

```python
cleaned_text = []
for txt in df.text:
    word_tokens = word_tokenize(txt)  
    filtered_sentence = ' '.join([w for w in word_tokens if not w in stop_words]  )
    cleaned_text.append(filtered_sentence)
cleaned_text[0:5]
```

```python
# 5 Word stemming -> helpful in text analysis and classificaion due to standardisation
cleaned_text = []
for txt in df.text:
    word_tokens = word_tokenize(txt)  
    porter = PorterStemmer()
    stemmed = ' '.join([porter.stem(word) for word in word_tokens])
    cleaned_text.append(stemmed)
cleaned_text[0:5]
```

Frequency of mentions of hashtags, meaningful text term(s): Education and Most used @ mentions.

```python
term = "education" #define term to pass a parameter for this section and the next
def getHashTags(data):
    term_freq = 0
    hastage_dic_count = {}
    mentions_dic_count = {}

    tweets = data.text
    for tweet in tweets:
        words = tweet.split(" ")
        for word in words:
            if term in word:
                term_freq +=1
                print(word)
            word = word.replace("\n","")
            if "#" in word and "http" not in word: #ensures the result does not return the term if it is in an URL
                hastage_dic_count[word] = hastage_dic_count.get(word, 0) + 1
            if "@" in word:
                mentions_dic_count[word] = mentions_dic_count.get(word, 0) + 1

    return hastage_dic_count, term_freq, mentions_dic_count
# here all hashtags are printed (not only one) with the number of times this hashtag is mentioned

hashtags_count, term_term, mentions_dic_count = getHashTags(df)
for k in sorted(hashtags_count, key=hashtags_count.get, reverse=True):
    print(k, hashtags_count[k]) 
```
